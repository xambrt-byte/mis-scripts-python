```
import matplotlib.pyplot as plt
from numpy import arange, sqrt
# LIENZO
plt.figure(figsize=(7,7))
plt.xlim(-4,12)
plt.ylim(-7,9)
plt.xticks(arange(-4,13,1), color='darkblue')
plt.yticks(arange(-7,10,1), color='darkblue')
plt.grid()
plt.title('GEOMETRIA BASICA', color='r')

#RECTAS
x1=arange(4,6,0.001)
y1=(1/2)*x1-1
plt.plot(x1,y1, 'darkgreen')
#RECTAS
x2=arange(2,4,0.001)
y2=(-1/2)*x2+3
plt.plot(x2,y2, 'darkgreen')
#RECTAS
x3=arange(3,4,0.001)
y3=(2/1)*x3-7
plt.plot(x3,y3, 'darkgreen')
#ELIPSE
x4=arange(1,7, 0.0001)
y5=-2-4*sqrt(1-(x4-4)**2/9)
plt.plot(x4,y5, 'darkblue')
#ELIPSE
x6=arange(1,7, 0.0001)
y7=4+4*sqrt(1-(x6-4)**2/9)
plt.plot(x6,y7, 'darkblue')
# ELIPSE
x8=arange(6,11,0.0001)
y8=1-3*sqrt(1-(x8-7)**2/16)
y9=1+3*sqrt(1-(x8-7)**2/16)
plt.plot(x8,y8,'deepskyblue')
plt.plot(x8,y9,'deepskyblue')
# ELIPSE
x10=arange(-3,2,0.0001)
y10=1-3*sqrt(1-(x10-1)**2/16)
y11=1+3*sqrt(1-(x10-1)**2/16)
plt.plot(x10,y10,'deepskyblue')
plt.plot(x10,y11,'deepskyblue')
# CIRCUNFERENCIA
x12= arange(1,3,0.001)
y12=2+sqrt(1-(x12-2)**2)
y13=2-sqrt(1-(x12-2)**2)
plt.plot(x12,y12, 'r')
plt.plot(x12,y13, 'r')
# CIRCUNFERENCIA
x14=arange(2,4,0.001)
y14=-1+sqrt(1-(x14-3)**2)
y15=-1-sqrt(1-(x14-3)**2)
plt.plot(x14,y14, 'r')
plt.plot(x14,y15, 'r')
# CIRCUNFERENCIA
x16=arange(5,7,0.001)
y16=2+sqrt(1-(x16-6)**2)
y17=2-sqrt(1-(x16-6)**2)
plt.plot(x16,y16, 'r')
plt.plot(x16,y17, 'r')
# PARABOLA
x18 = arange(0,8,0.001)
y18 = 5 - (1/4)*(x18 - 4)**2
plt.plot(x18, y18, 'darkorange')
# PARABOLA
x19 = arange(0,8,0.001)
y19 = -3 + (1/4)*(x19 - 4)**2
plt.plot(x19, y19, 'darkorange')

plt.plot(x19, y19, 'darkorange')
plt.axis('equal')
plt.show()
```
