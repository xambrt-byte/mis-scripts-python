```
from numpy import *

A = array([[-2,-3,1], [2,2,-1], [-3,-1,3]])
B = array([[2,3,-3,2], [3,2,-1,2], [2,-2,3,-2]])

C = (A.T @ B) - 2*B
print('(A.T @ B) - 2*B:\n', C)
```

```
from numpy import arange, array, sqrt
import matplotlib.pyplot as plt

# Lienzo
plt.figure(figsize=(6, 5))
plt.xlim(-10, 10)
plt.ylim(-8, 8)
plt.xticks(arange(-10, 11, 1))
plt.yticks(arange(-8, 9, 1))
plt.axhline(color='k')
plt.axvline(color='k')
plt.grid()

# Construir una nave con puntos
nave = array([[3,-3,-1,-1,-3,3],[0,-3,-1,1,3,0],[1,1,1,1,1,1]])
x = nave[0]; y = nave[1]; plt.plot(x, y, 'b')

# Definir desplazamiento con dx y dy
dx = 7; dy = 3
xx = [0,dx]; yy = [0,dy]
plt.plot(xx, yy, 'r--')

# Rotación y Traslación definida por dx y dy
senx = dy/sqrt(dx**2+dy**2)
cosx = dx/sqrt(dx**2+dy**2)
rotars = array([[cosx,-senx,dx],[senx,cosx,dy],[0,0,1]])
nave1 = rotars @ nave
# Gráfica de la rotación y traslación
x1 = nave1[0]
y1 = nave1[1]

plt.plot(x1, y1, 'b')

plt.show()
```

```
from numpy import arange, array, sqrt
import matplotlib.pyplot as plt

# Lienzo
plt.figure(figsize=(6, 5))
plt.xlim(-10, 10)
plt.ylim(-8, 8)
plt.xticks(arange(-10, 11, 1))
plt.yticks(arange(-8, 9, 1))
plt.axhline(color='k')
plt.axvline(color='k')
plt.grid()

# Construir una nave con puntos
nave = array([[3, -3, -1, -1, -3, 3], [0, -3, -1, 1, 3, 0], [1, 1, 1, 1, 1, 1]])
x = nave[0]
y = nave[1]
plt.plot(x, y, 'b')

# Definir desplazamiento con dx y dy
dx = 7
dy = 3
xx = [0, dx]
yy = [0, dy]
plt.plot(xx, yy, 'r--')

# Rotación y Traslación para el cuadrante 1
senx = dy / sqrt(dx**2 + dy**2)
cosx = dx / sqrt(dx**2 + dy**2)
rotars = array([[cosx, -senx, dx], [senx, cosx, dy], [0, 0, 1]])
nave1 = rotars @ nave

# Gráfica de la rotación y traslación
x1 = nave1[0]
y1 = nave1[1]
plt.plot(x1, y1, 'g')


dx2 = -2.5
dy2 = 4.4
xx2 = [0, dx2]
yy2 = [0, dy2]
plt.plot(xx2, yy2, 'r--')

# Rotación y Traslación
senx2 = dy2 / sqrt(dx2**2 + dy2**2)
cosx2 = dx2 / sqrt(dx2**2 + dy2**2)
rotars2 = array([[cosx2, -senx2, dx2], [senx2, cosx2, dy2], [0, 0, 1]])
nave2 = rotars2 @ nave

# Gráfica de la rotación y traslación
x2 = nave2[0]
y2 = nave2[1]
plt.plot(x2, y2, 'y')

dx3 = -2.5
dy3 = -4.4
xx3 = [0, dx3]
yy3 = [0, dy3]
plt.plot(xx3, yy3, 'r--')

# Rotación y Traslación
senx3 = dy3 / sqrt(dx3**2 + dy3**2)
cosx3 = dx3 / sqrt(dx3**2 + dy3**2)
rotars3 = array([[cosx3, -senx3, dx3], [senx3, cosx3, dy3], [0, 0, 1]])
nave3 = rotars3 @ nave

# Gráfica de la rotación y traslación
x3 = nave3[0]
y3 = nave3[1]
plt.plot(x3, y3, 'c')


dx4 = 7
dy4 = -3
xx4 = [0, dx4]
yy4 = [0, dy4]
plt.plot(xx4, yy4, 'r--')

# Rotación y Traslación
senx4 = dy4 / sqrt(dx4**2 + dy4**2)
cosx4 = dx4 / sqrt(dx4**2 + dy4**2)
rotars4 = array([[cosx4, -senx4, dx4], [senx4, cosx4, dy4], [0, 0, 1]])
nave4 = rotars4 @ nave

# Gráfica de la rotación y traslación
x4 = nave4[0]
y4 = nave4[1]
plt.plot(x4, y4, 'm')

plt.show()
```
