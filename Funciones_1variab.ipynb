```
# @title
from numpy import arange, sin, pi
import matplotlib.pyplot as plt
plt.figure(figsize=(12,9))
plt.xlim(-4,8)
plt.ylim(-1,8)
plt.xticksx = arange(-4,8,1)
plt.yticksy = arange(-1,9,1)
plt.grid()
plt.title("FUNCIONES DE UNA VARIABLE")
#FUNCION SENO 1
x1=arange(-3,3.1,0.1)
y1=sin((pi/4)*(x1+3))+4
plt.plot(x1,y1,"bo")
#FUNCION SENO 2
x2=arange(-3,3.1,0.1)
y2=sin((pi/4)*(x2-3))+3
plt.plot(x2,y2,"bo")
#RECTA 1
x3=arange(3,6.1,0.1)
y3=((1/3)*x3)+2
plt.plot(x3,y3,"ro")
#RECTA 2
x4=arange(1,4.1,0.1)
y4=(-2/3)*x4+(8/3)
plt.plot(x4,y4,"ro")
#funcion 5
x5=arange(4,6.1,0.1)
y5=-x5**2+12*x5-32
plt.plot(x5,y5,"ko")
#funcion 6
x6=arange(0.3,7.1,0.1)
y6=-0.245*x6**2+1.715*x6+4
plt.plot(x6,y6,"ko")
#funcion 7
x7=arange(1,7.1,0.1)
y7=-((x7**2)/9)+((8*x7)/9)+(29/9)
plt.plot(x7,y7,"yo")

plt.plot(-1,4, "go", markersize=10)
```
