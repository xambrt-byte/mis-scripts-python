```
import matplotlib.pyplot as plt
from numpy import arange,exp
plt.figure(figsize=(6,4))
plt.xlim(-6,6)
plt.ylim(-0.2,1.2)
plt.xticks(arange(-6,7,1))
plt.yticks(arange(-0.2,1.4,0.2))
plt.title("Funciones sigmoide")
plt.grid()
x1=arange(-6,6.1,0.1)
y1=1/(1+exp(-1*x1))
plt.plot(x1,y1,"r")
x2=arange(-6,6.1,0.1)
y2=1/(1+exp(-2*x2))
plt.plot(x2,y2,"g")
x3=arange(-6,6.1,0.1)
y3=1/(1+exp(-4*x3))
plt.plot(x3,y3,"b--")
plt.show()
```
