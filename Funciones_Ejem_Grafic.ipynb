```
from sympy import *
from numpy import arange
import matplotlib.pyplot as plt

x=symbols("x")
y=0.1*x**5-2.1*x**4+16*x**3-54*x**2+78.4*x-26.4

print("Maximos y minimos")
y1=diff(y,x,1)
y2=diff(y,x,2)
pcrit=[(xx,y.subs(x,xx)) for xx in solveset(y1,x)]
x_pcrit, y_pcrit = zip(*pcrit)
for i in range(0,len(pcrit)):
  der2da= y2.subs(x,x_pcrit[i])
  if der2da > 0 :
    print("P",i+1,":",pcrit[i], "Min")
  else:
    print("P", i+1,":",pcrit[i],"Max")
    print("PUNTO DE INFLEXION")
pinfl=[(xx,y.subs(x,xx)) for xx in solveset(y2,x)]
x_pinfl,y_pinfl=zip(*pinfl)
for i in range(0,len(pinfl)):
  print("P",i+1,"i",pinfl[i])

x1=arange(1,8.1,0.01)
y1=0.1*x1**5-2.1*x1**4+16*x1**3-54*x1**2+78.4*x1-26.4
plt.plot(x1,y1,"k")
plt.grid()

plt.plot(x_pcrit, y_pcrit,"bo")
plt.plot(x_pinfl,y_pinfl,"mx")
plt.title("Analisis de funciones")
plt.show()
```

```
from sympy import *
import numpy as np
import matplotlib.pyplot as plt

plt.figure(figsize=(6,6))
plt.xlim(-2,2)
plt.ylim(-6,4)
plt.xticks(np.arange(-2,3,1))
plt.yticks(np.arange(-6,5,1))
plt.grid()

x=symbols("x")
func=-2*sin(x)+2*cos(2*x)

taylor_series= series(func,x0=0,n=8).removeO()

taylor_func=lambdify(x,taylor_series,"numpy")

x1=np.arange(-2,2.01,0.01)
y1= -2*np.sin(x1)+2*np.cos(2*x1)
y_taylor= taylor_func(x1)
plt.title("MODELO DE APROXIMACION DE TAYLOR")
plt.plot(x1,y1,"b",label="Funcion original")
plt.plot(x1,y_taylor,"k--",label="Funcion aprox. de taylor")
plt.legend()

plt.show()
print("Modelo de aproximacion\n")
taylor_series
```

```
from numpy import *
A=array([[-2,4,3],[3,5,-1],[4,3,2],[2,-5,1]])
B=array([[3,-2,4,5],[0,2,3,-1],[7,5,-3,-1]])
res=(2*A.T)-3*B
res1=(2*B.T-A)@A.T
res2=((A@B)@B.T)
print("2*A.T-3*B =\n",res)
print("(2*B.T-A)@A.T =\n",res1)
print("(A@B)@B.T =\n",res2)
```

```
from numpy import *
x=15
A= array([[cos(x),-sin(x),0],[sin(x),cos(x),0],[0,0,1]])
A_inversa= linalg.inv(A)
ort= A@A_inversa
print("A@A_inversa= \n",round(ort))
print("La matriz A es ortogonal")
```
