**1 REPRESENTACION DE VECTORES**


```
from numpy import *
from numpy.linalg import *
from sympy import *

u = array([-2,5,-1,2,4,8])
print('vector u=',u)
print('dimension de u=', u.shape)
```

2 CARCTERISTICAS DE VECTOORES

```
#NORMA L1 Y NORMA L2
v=array([6,3,-2])
print('vector v=', v)
norma_l1=norm(v, 1)
norma_l2=norm(v, 2)
print('Norma L1=', norma_l1)
print('Norma L2=', norma_l2)
```

```
# VECTOR UNITARIO=VECTOR/NORMA EUCLIDIANA
unit_v=v/norma_l2
print('Vector unitario de v=',unit_v)
```

3 OPERSCIONES

```
#sumas y restas
u=array([3,2,-5])
v=array([-5,3,1])
print('vector u=', u)
print('vector v=',v)
print('suma u+v=',u+v)
print('resta u-v',u-v)
```

4 multiplicacion

```
# multiplicacion de un escalar por un vector
print('vector u=', u)
print('vector v=', v)
print('multiplicaion 3*v=',3*v)
```

5 PRODUCTO DE UN PUNTO

```
#producto punto
print('vector u=', u)
print('vector v=', v)
print('producto punto u-v=', dot(u,v))
print('producto punto u-v=', u.dot(v))
print('producto punto u-v=',u @ v)
```

1) INDEPENDEMCIA LINEAL

```
#MATRIZ REDUCIDA ESCALONADA
from sympy import Matrix, pprint
M = Matrix([ [-2,1,1,-1],[1,1,-1,3],[3,-3,-1,-2] ])
print('Matriz de vectores columna:\n')
pprint(M)
M_re = M.rref()
print ('Matriz reducida escalonada:')
pprint(M_re)
```

```
import numpy as np

u = np.array([1, 2])
v = np.array([3, 4])

print('vector u=', u)
print('vector v=', v)

# CAMBIA ESTO:
# norma_u = norm(u)
# POR ESTO:
norma_u = np.linalg.norm(u)
norma_v = np.linalg.norm(v)

# También corrige las mayúsculas (norma_U debe ser norma_u)
# y usa np.rad2deg y np.arccos
angulo_uv = np.rad2deg(np.arccos((u @ v) / (norma_u * norma_v)))

print('angulo entre u y v=', angulo_uv, 'en grados')
```
